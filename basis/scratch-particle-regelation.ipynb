```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
\frac{\partial z}{\partial t} = -\frac{A}{36\pi\mu a}\frac{1}{z}
$$

$$
\int_0^\zeta\frac{\partial z}{\partial t} = -\frac{A}{36\pi\mu a}ln(z)|_{z = 0}^{z = \zeta} 
$$

```python
A = 4.6e-21
mu = 1.8e-3
a = 1e-6
coeff = -A / (36 * np.pi * mu * a)

zs = np.arange(0.01, 5.01, 0.01)

plt.plot(coeff * np.log(zs), zs)
plt.axvline(0.0, linestyle = ':', color = 'k')
plt.show()
```

```python
A = 4.6e-21
mu = 1.8e-3
r = 5e-5
ki = 2
kp = 6
ρ = 917
L = 3.34e5
dT = 0.12
h = np.arange(1e-6, 1.0 + 1e-2, 1e-2)

β = A / (36 * np.pi * mu * r * (2 * ki + kp))
V = (3 * ki) / ((h / β) + (ρ * L / (dT / h)))

plt.plot(h[2:], V[2:])
```
