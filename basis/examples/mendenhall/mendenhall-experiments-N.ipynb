```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio

from landlab import RasterModelGrid
from landlab.plot import imshow_grid

import sys
sys.path.append('/home/egp/repos/basal-ice-facies/basis/src/')
from src.basal_ice_stratigrapher import BasalIceStratigrapher
```

```python
# Parameters
secpera = 31556926
sliding_scalar = 0.6
```

### Start: effective pressure = 0.05 * P$_i$

```python
N_scales = [0.05, 0.1, 0.2, 0.4]
models = {i: BasalIceStratigrapher('inputs/input_file.toml') for i in range(len(N_scales))}

with rio.open('./inputs/ice_thickness.tif', 'r') as f:
    data = np.flip(f.read(1), axis = 0)
    
    for key, model in models.items():
        model.grid.add_field('glacier__thickness', data, at = 'node', clobber = True)
        model.grid.add_field('glacier__effective_pressure', 
                             model.grid.at_node['glacier__thickness'][:] * 917 * 9.81 * N_scales[key], 
                             at = 'node', clobber = True)
        
with rio.open('./inputs/surface_velocity_x.tif', 'r') as f:
    data = -1 * np.flip(f.read(1), axis = 0) * sliding_scalar * (1/secpera)
    
    for key, model in models.items():
        model.grid.add_field('glacier__sliding_velocity_x', data, at = 'node', clobber = True)
    
with rio.open('./inputs/surface_velocity_y.tif', 'r') as f:
    data = np.flip(f.read(1), axis = 0) * sliding_scalar * (1/secpera)
    
    for key, model in models.items():
        model.grid.add_field('glacier__sliding_velocity_y', data, at = 'node', clobber = True)

for key, model in models.items():
    model.grid.add_field('glacier__sliding_velocity', 
                         np.sqrt(model.grid.at_node['glacier__sliding_velocity_x'][:]**2 + model.grid.at_node['glacier__sliding_velocity_y'][:]**2), 
                         at = 'node', clobber = True)
    model.calc_erosion_rate()
    model.calc_melt_rate()
```

```python
fig, ax = plt.subplots(1, 4, figsize = (30, 6))

for key, model in models.items():
    
    im = ax[key].imshow(np.flip(np.reshape(model.grid.at_node['glacier__effective_pressure'][:] * 1e-3, model.grid.shape), axis = 0),
                   cmap = 'pink')
    plt.colorbar(im, ax = ax[key])
    ax[key].set_title('N = ' + str(N_scales[key]) + ' * P$_i$ (kPa)')
    
    
plt.show()
```

```python
# CFL calculation
C = 1/12
dt = C / (np.max(models[0].grid.at_node['glacier__sliding_velocity_x'][:] / models[0].grid.spacing[0]) + 
          np.max(models[0].grid.at_node['glacier__sliding_velocity_y'][:] / models[0].grid.spacing[1]))
print(dt)

secpera / dt
```

```python
dt = 60 * 60 * 24

for key, model in models.items():
    
    for i in range(100):
        model.run_one_step(i * 1e-2, advection = False, dynamic_thinning = False)
        
    for i in range(100):
        model.run_one_step(i, advection = False, dynamic_thinning = False)
        
    for i in range(100):
        model.run_one_step(100, advection = False, dynamic_thinning = False)
        
    print('Spin-up completed for model #' + str(key + 1))
   
    for i in range(100):
        for j in range(73):
            model.run_one_step(dt * 5, advection = False, dynamic_thinning = False)
        
        if i % 10 == 0:
            print('Year ' + str(i) + ' completed for model #' + str(key + 1))
```
