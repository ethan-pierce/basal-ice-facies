```python
import numpy as np
import matplotlib.pyplot as plt

from landlab import RasterModelGrid
from landlab.plot import imshow_grid
```

```python
Lx = 15e3
Lz = 750
dx = 30
dz = 1.5
```

```python
print('shape: (' + str(Lx / dx) + ', ' + str(Lz / dz) + ')')
```

```python
print('aspect ratio: ' + str(Lx / Lz))
```

```python
grid = RasterModelGrid((Lx / dx, Lz / dz), (dx, dz))
```

```python
fig, ax = plt.subplots(figsize = (15, 3))

xs = np.arange(Lx)
curve_parameter_a = -1 / (140625)
curve_parameter_b = 7 / 75
curve_parameter_c = 300
ice_surface = curve_parameter_a * xs**2 + curve_parameter_b * xs + curve_parameter_c

ax.plot(xs, ice_surface)
plt.show()
```

```python
glacier = np.empty(grid.number_of_nodes)
surface = np.empty(grid.number_of_nodes)

parametric_a = -1 / (140625)
parametric_b = 7 / 75
parametric_c = 300

for node in range(grid.number_of_nodes):
    x_coordinate = int(grid.node_x[node])
    z_coordinate = int(grid.node_y[node])
    
    surface[node] = parametric_a * x_coordinate**2 + parametric_b * x_coordinate + parametric_c
    
    if z_coordinate < surface[node]:
        glacier[node] = 1
    else:
        glacier[node] = 0
```

```python
fig, ax = plt.subplots(figsize = (15, 3))

ax.imshow(np.flip(glacier.reshape(grid.shape), 0), cmap = 'Blues_r', aspect = 0.25)
plt.show()
```

```python
m = -0.00933
b = 150
decay = 0.6
vel_a = -7 / 2812500
vel_b = 14 / 375
vel_c = 10

velocity_x_dy = glacier * ((1 - decay) * (grid.node_y / surface) + decay)
velocity_x_dx = glacier * (vel_a * grid.node_x**2 + vel_b * grid.node_x + vel_c)
velocity_x = velocity_x_dy * velocity_x_dx
```

```python
fig, ax = plt.subplots(figsize = (15, 3))

im = ax.imshow(np.flip(velocity_x.reshape(grid.shape), 0), cmap = 'Spectral_r', aspect = 0.25)
plt.colorbar(im)
plt.show()
```

```python
diff_ux = grid.calc_diff_at_link(velocity_x)
gradient_ux = grid.map_mean_of_horizontal_active_links_to_node(diff_ux)
velocity_z = -gradient_ux * grid.node_y
```

```python
fig, ax = plt.subplots(figsize = (15, 3))

im = ax.imshow(np.flip(gradient_ux.reshape(grid.shape), 0), cmap = 'Spectral_r', aspect = 0.25)
plt.colorbar(im)
plt.show()
```

```python
def pressure_melting_point(depth):
    clapeyron = 7.42e-8
    tref = 273.16
    pref = 611.73
    rho = 917
    g = 9.81
    pressure = rho * g * depth
    tm = tref - clapeyron * (pressure - pref)
    return tm
```

```python
def calc_temperature(enthalpy, depth):
    rho = 917
    heat_cap = 2010
    tref = 273.16
    tmelt = pressure_melting_point(depth)
    
    temperate = True if h < (rho * heat_cap * (tmelt - tref)) else False
    
    if temperate:
        temp = tmelt
    else:
        temp = tref + h / (rho * heat_cap)
        
    return temp
```

```python
def calc_porosity(enthalpy, depth):
    rho = 917
    heat_cap = 2010
    tref = 273.16
    tmelt = pressure_melting_point(depth)
    rho_w = 1000
    latent = 3.34e5
    
    temperate = True if h < (rho * heat_cap * (tmelt - tref)) else False
    
    if temperate:
        porosity = (enthalpy - (rho * heat_cap * (tmelt - tref))) / (rho_w * latent)
    else:
        porosity = 0
        
    return porosity
```

```python
def calc_enthalpy(temperature, porosity):
    rho = 917
    heat_cap = 2010
    tref = 273.16
    rho_w = 1000
    latent = 3.34e5
    
    enthalpy = rho * heat_cap * (temperature - tref) + rho_w * latent * porosity
    return enthalpy
```

```python
pmp = glacier * pressure_melting_point(grid.node_y)
initial_enthalpy = glacier * calc_enthalpy(pmp - 1, 0)
```

```python
im = plt.imshow(np.flip(pmp.reshape(grid.shape), 0), cmap = 'Spectral_r', aspect = 0.25, vmin = 271, vmax = 273.16)
plt.colorbar(im)
plt.show()
```
