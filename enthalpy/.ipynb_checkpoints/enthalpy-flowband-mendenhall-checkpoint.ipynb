```python
import numpy as np
import matplotlib.pyplot as plt

%load_ext pyinstrument
import time

%load_ext autoreload
%aimport src.enthalpy_flowband
%autoreload 1
from src.enthalpy_flowband import BoundaryCondition, EnthalpyFlowbandState, EnthalpyFlowbandExplicit
```

```python
def MH_surface(xs):
    a = -1 / 140625
    b = 7 / 75
    c = 300
    
    return a * xs**2 + b * xs + c

def MH_velocity(xs, zs, surface):
    a = -7 / 2812500
    b = 14 / 375
    c = 10
    decay = 0.6
    shear_ratio = 1 - decay
    fn_x = a * xs**2 + b * xs + c
    fn_z = (1 - decay) * (zs / surface) + decay
    
    return fn_x * fn_z * 3.14e-8

def MH_temperature(t):
    a = 16
    f = 1 / 3.14e7
    b = 20
    
    return a * np.sin(2 * np.pi * f * t) + b
```

```python
sec_per_a = 3.14e7
total_a = 100
t_final = sec_per_a * total_a
dt = 1e6
```

```python
state = EnthalpyFlowbandState(500, 500, 30, 1.5, t_final, dt)
```

```python
state.set_initial_surface(MH_surface)
state.set_pressure_melting_point()
state.set_initial_velocity(MH_velocity)
state.set_air_temperature(MH_temperature)
state.calc_internal_energy()
state.set_initial_conditions(initial_temperature = 268)
```

```python
to_plot = np.reshape(state.in_glacier, (state.nx, state.nz)) * np.gradient(np.reshape(state.porosity, (state.nx, state.nz)), axis = 1)

fig, ax = plt.subplots(figsize = (15, 3))
im = ax.imshow(np.flip(to_plot.reshape(state.grid.shape), 0), cmap = 'Spectral_r', aspect = 0.25)
plt.colorbar(im)
plt.show()
```

```python
temp_in = np.full(state.nz, 253)
temp_out = np.full(state.nz, 253)
temp_surface = np.full(state.nx, 253)
temp_base = np.full(state.nx, state.melt_point[0:state.nx])
temperature_bc = BoundaryCondition('temperature', temp_in, temp_out, temp_surface, temp_base)
```

```python
por_in = np.full(state.nz, 0.0)
por_out = np.full(state.nz, 0.0)
por_surface = np.full(state.nx, 0.0)
por_base = np.full(state.nx, 0.0)
porosity_bc = BoundaryCondition('temperature', por_in, por_out, por_surface, por_base)
```

```python
CFL = 0.1
model = EnthalpyFlowbandExplicit(state, CFL, temperature_bc, porosity_bc)
model.identify_boundaries(MH_surface)
model.inflow_bc = model.calc_local_enthalpy(temp_in, por_in)
model.outflow_bc = model.calc_local_enthalpy(temp_out, por_out)
model.top_bc = model.calc_local_enthalpy(temp_surface, por_surface)
model.bottom_bc = model.calc_local_enthalpy(temp_base, por_base)
```

```python
%%pyinstrument
model.run_one_step(1)
```
